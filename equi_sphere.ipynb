```python
import math
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
N = 32 # numnber of points on sphere
r = 10 # radius

# Algo 1
z = [-r + i*(2*r)/N for i in range(N+1)]
phi = [i*(2*math.pi)/N for i in range(N+1)]
x = [np.sqrt(r**2 - a**2)*math.cos(b) for a,b in zip(z,phi)]
y = [np.sqrt(r**2 - a**2)*math.sin(b) for a,b in zip(z,phi)]
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
N = 32 # numnber of points on sphere
r = 1 # radius

a = 4*math.pi*r**2/N
d = np.sqrt(a)
Mv = int(np.round(math.pi/d))
dv = math.pi/Mv

xs = []
ys = []
zs = []
for i in range(Mv):
    v = math.pi*(i + 0.5)/Mv
    Mphi = int(np.round(2*math.pi*math.sin(v)/dv))
    for j in range(Mphi):
        phi = 2*math.pi*j/Mphi
        x = r*math.sin(v)*math.cos(phi)
        y = r*math.sin(v)*math.sin(phi)
        z = r*math.cos(v)
        
        xs.append(x)
        ys.append(y)
        zs.append(z)
        
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs, ys, zs, c='r', marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_xlim([-r,r])
ax.set_ylim([-r,r])
ax.set_zlim([-r,r])
plt.show()
```
